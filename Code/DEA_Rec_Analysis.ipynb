```python
# import dependencies 
import os 
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns 
# check the current directory and change if need be
os.getcwd()
# read in data for file that was manually produced in Excel
df=pd.read_csv("DEA_Model91_Rec.csv")
```

```python
####################################    The meaning of each feature in this file      ####################################################
#Station: The name of station
#Quarter: Which of the 13 Quarters 
#IRF_forms: The number of IRF forms collected for station, quarter pair
#IRF_Com:The IRF form completion score for the collected station, quarter pair
#VIF_forms:The number of VIF forms collected for station, quarter pair
#VIF_Com:The VIF form completion score for the collected station, quarter pair
#IRF_forms_Rec: The reccomended increase needed for IRF forms collected  for the station, quarter pair to be seen as efficient
#IRF_Com_Rec: The reccomended increase needed for IRF form completion for the station, quarter pair to be seen as efficient
#VIF_forms_Rec: The reccomended increase needed for VIF forms collected  for the station, quarter pair to be seen as efficient
#VIF_Com_Rec: The reccomended increase needed for VIF form completion for the station, quarter pair to be seen as efficient
#Actual_VIF: The actual observed number of VIF forms for each station, quarter pair *not used in this file*
#Actual_IRF:The actual observed number of IRF forms for each station, quarter pair *not used in this file*
#Actual_VIFCom:The actual observed VIF form completeness for each station, quarter pair *not used in this file*
#Actual_IRFCom:The actual observed VIF form completeness for each station, quarter pair *not used in this file*
#VIF_Perc_Increase: The recommended percentage increase for the number of VIF forms for each station, quarter pair
#IRF_Perc_Increase: The recommended percentage increase for the number of IRF forms for each station, quarter pair
#VIFCom_Perc_Increase: The recommended percentage increase for the VIF form completion for each station, quarter pair
#IRFCom_Perc_Increase: The recommended percentage increase for the IRF form completion for each station, quarter pair
```

```python
# create the feature to turn the form completeness to percentages
df['IRF_Com']=df['IRF_Com']*100
df['VIF_Com']=df['VIF_Com']*100
```

```python
# Creates Figure 6: The Additional Percentage of IRF Forms Needed for Efficiency Varies by Quarter and Station.
fig,axes=plt.subplots(figsize=(8,4))
sns.scatterplot(data=df[df['IRF_Perc_Increase']>=0], x='Quarter',y='Station', size='IRF_Perc_Increase',color='black')
legend=plt.legend(title='  Additional \n Percentage',bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

for i, text in enumerate(legend.get_texts()):
    text.set_text(f"{i}%")

plt.title('IRF Forms: Additional Percentage Required for Efficiency')
axes.set_xlim(0, 14)
axes.set_xticks(np.arange(1, 14, step=1))
plt.tight_layout()
plt.savefig("IRF_Forms_Recs_inpercent.png")
```

```python
# Creates Figure 7: The Additional Percentage of VIF Forms Needed for Efficiency Varies by Quarter and Station.
fig,axes=plt.subplots(figsize=(8,4))
sns.scatterplot(data=df[df['VIF_Perc_Increase']>=0], x='Quarter',y='Station', size='VIF_Perc_Increase',color='black')
legend=plt.legend(title='  Additional \n Percentage',bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
for i, text in enumerate(legend.get_texts()):
    text.set_text(f"{i}%")
plt.title('VIF Forms: Additional Percentage Required for Efficiency')
axes.set_xlim(0, 14)
axes.set_xticks(np.arange(1, 14, step=1))
plt.tight_layout()
plt.savefig("VIF_Forms_Recs_inpercent.png")
```

```python
# Creates Figure 8: The  Additional  Percentage  of  IRF  Form  Completion  Needed  for  Efficiency  Varies  by  Quarter  and Station.
fig,axes=plt.subplots(figsize=(8,4))
sns.scatterplot(data=df[df['IRFCom_Perc_Increase']>=0], x='Quarter',y='Station', size='IRFCom_Perc_Increase',color='black')
legend=plt.legend(title='  Additional \n Percentage',bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
for i, text in enumerate(legend.get_texts()):
    text.set_text(f"{i}%")
plt.title('IRF Completion: Additional Percentage Required for Efficiency')
axes.set_xlim(0, 14)
axes.set_xticks(np.arange(1, 14, step=1))
plt.tight_layout()
plt.savefig("IRF_Comp_Recs_inpercent.png")
```

```python
# Creates Figure 9: The  Additional  Percentage  of  VIF  Form  Completion  Needed  for  Efficiency  Varies  by  Quarter  and Station.
fig,axes=plt.subplots(figsize=(8,4))
sns.scatterplot(data=df[df['VIFCom_Perc_Increase']>=0], x='Quarter',y='Station', size='VIFCom_Perc_Increase',color='black')
legend=plt.legend(title='  Additional \n Percentage',bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
for i, text in enumerate(legend.get_texts()):
    text.set_text(f"{i}%")
plt.title('VIF Completion: Additional Percentage Required for Efficiency')
axes.set_xlim(0, 14)
axes.set_xticks(np.arange(1, 14, step=1))
plt.tight_layout()
plt.savefig("VIF_Comp_Recs_inpercent.png")
```
